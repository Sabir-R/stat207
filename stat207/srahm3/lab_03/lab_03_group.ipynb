### Renaming a Column

We can rename a subset of column names by using the **.rename()** function and giving the **columns** parameter within it a dictionary in which the **keys** of the dictionary are the old column names we would like to change, and the corresponding **values** in the dictionary are the corresponding column names that we would like to change them to.


```python
import pandas as pd
tmp1 = pd.DataFrame({'a': ['Nick','Kevin','Joe'], 'b': [4,5,6]})
tmp1
```

```python
tmp1=tmp1.rename(columns = {'a': 'c', 'b': 'd'})
tmp1
```

### Outer Join Merge

If we would like to join our dataframes based on a primary key (or set of primary keys) values AND we would like to keep all rows from both dataframe, not just the ones that have a primary key match, we can specify that we would like to perform an **outer join** by setting the **how='outer'** parameter.

Notice how the rows from tmp1 and tmp2 that did not have a primary key match are given NaN values for the dataframe values that did not have any corresponding values for it.

```python
tmp2 = pd.DataFrame({'c': ['Nick','Kevin','Mildred'], 'b': [7,8,9]})
tmp2
```

```python
tmp3 = pd.merge(tmp1, tmp2, how='outer')
tmp3
```

### Filling in Missing Values

Rather than dropping rows with missing values, we can fill them in with a number that we supply by using the **.fillna()** function. For instance, in the code below we fill in all NaN values with a 0.

```python
tmp3 = tmp3.fillna(0)
tmp3
```

### Creating New Variables with Mathematical Expressions

We can perform mathematical operations on dataframe columns to create new dataframe columns.

```python
tmp3['difference'] = tmp3['d'] - tmp3['b']
tmp3
```

## 1. Data Processing

First, let's load and process this dataset.

### 1.1 Read the csv
First, read the **steam_sample_cleaned.csv** file into a dataframe and display the first five rows of the dataframe.

**Note: The data in this 'steam_sample_cleaned.csv' file is the same data that we used in lab 2, except we have already cleaned and dropped all rows with missing values. So you do not need to perform any data cleaning in this assignment.**

```python
df = pd.read_csv('steam_sample_cleaned.csv')
df.head()
```

### 1.2. Separate the dataframes. 

Again, in this analysis we would like to analyze the game purchase information and game play information separately. First create two new dataframes:
* one that is comprised of the purchase rows
* the other that is comprised of the play rows.

```python
df_purchase = df[df['activity'] == 'purchase']
df_purchase = df_purchase.reset_index(drop=True)
df_purchase.head()
```

```python
df_play = df[df['activity'] == 'play']
df_play = df_play.reset_index(drop=True)
df_play.head()
```


## 2. <u> Calculating Popularity Metric 4</u>: Percent of Purchased Games Actually Played

A metric we can use to measure how popular a game is to find out what percent of users that purchased this game actually played this game.

### 2.1. Counts Dataframe

First, create a dataframe that that counts up how many times each game has been purchased. Your final dataframe should have the following format.

1. It should be comprised of just two columns.
    * One column should be game name. This column should be named 'game_name'.
    * the other column should be the number of times this respective game has been purchased. This column should be named 'purchase_counts'.
2. It should have an index going from 0,1,2... etc.

```python
df_purchase_count = df_purchase.groupby('game_name').agg('sum').reset_index()
df_purchase_count = df_purchase_count.iloc[:,[0,2]]
df_purchase_count = df_purchase_count.rename(columns = {'hours_played_if_play': 'purchase_counts'})
df_purchase_count = df_purchase_count.sort_values("purchase_counts",ascending=False).reset_index(drop=True)

df_purchase_count.head()
```

### 2.2. Play Counts Dataframe

Next, create a dataframe that that counts up how many players have played this game for some amount of time. Your final dataframe should have the following format.

1. It should be comprised of just two columns.
    * One column should be game name. This column should be named 'game_name'.
    * the other column should be the number of times this respective game has been played. This column should be named 'play_counts'.
2. It should have an index going from 0,1,2... etc.

```python
df_play_time = df_play.groupby('game_name').agg('sum').reset_index()
df_play_time = df_play_time.iloc[:,[0,2]]
df_play_time = df_play_time.rename(columns = {'hours_played_if_play': 'play_counts'})
df_play_time = df_play_time.sort_values("play_counts",ascending=False).reset_index(drop=True)

df_play_time.head()
```

### 2.3. Outer Join Merging

Next, perform an outer-join merge between these two dataframe that you created in 2.1 and 2.2 on the *game_name* column.

```python
df_merged = pd.merge(df_purchase_count, df_play_time, how='outer')
df_merged
```

### 2.4. Filling in Missing Values with 0

Fill in all missing values in this dataframe from 2.3 with a 0.

```python
df_merged = df_merged.fillna(0)
df_merged
```

### 2.5. Calculating Popularity Metric 4 for Each Game

Finally, create a new column in this dataframe from 2.4 which calculates popularity metric 4 (ie. the # of users that played the game/# of users that purchased the game).

```python
df_merged['popularity metric'] = df_merged['play_counts'] / df_merged['purchase_counts']
df_merged
```
