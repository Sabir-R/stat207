## <u>Case Study</u>: Identifying and Exploring the Properties of Malignant Breast Tumors

In this analysis, we would like to build a classifier model that predicts whether a breast tumor is malignant or benign, based on it's tissue sample properties for *new* datasets. In addition, we would like to learn more about what tissue sample attribute properties (and attribute relationships) characterize malignant vs. benign breast tumor tissue samples.

The data in the **breast_tumor_scaled.csv** file contains a **random sample** of benign and malignant breast tumor tissue samples from 569 patients.


### More Dataset Information

The following physical attributes are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image. 

*Interesting Factoid: Furthermore, each of these physical attributes have already been **scaled**, so they should all roughly have the same standard deviation and mean. Scaling variables before running a regularized regression first can help the results not be dominated just by variables that have much larger scale.*

Specifically, the data in the **breast_tumor_scaled.csv** file contains the following attributes about these breast tumor tissue samples.

* **diagnosis** - whether the tumor was malignant ('M') or benign ('B')
* **radius_mean** 
* **texture_mean** 
* **perimeter_mean**
* **area_mean**
* **smoothness_mean** 
* **compactness_mean** 
* **concavity_mean** 
* **concave points_mean** 
* **symmetry_mean**
* **fractal dimension_mean**

### 3.1. Hypotheses

Hypothesis: There is an association between breast tumor diagnosis and tissue sample smoothness_mean
Alternative Hypothesis: There is not an association between breast tumor diagnosis and tissue sample smoothness_mean

### 3.2. Conditions

### 3.3. Test Statistic

Calculate the test statistic for this hypothesis test.

```python
df_benign = dfc[dfc['diagnosis'] == 'B'].reset_index(drop=True)
df_malignant = dfc[dfc['diagnosis'] == 'M'].reset_index(drop=True)
```

```python
dfbm = df_benign['smoothness_mean'].mean()
dfbstd = df_benign['smoothness_mean'].std()
dfmm = df_malignant['smoothness_mean'].mean()
dfmstd = df_malignant['smoothness_mean'].std()
```

```python
def tpr_fpr_thresh(y, pred_prob, thresh):
    yhat = 1*(pred_prob >= thresh)
    tn, fp, fn, tp = confusion_matrix(y_true=y, y_pred=yhat).ravel()
    tpr = tp / (fn + tp)
    fpr = fp / (fp + tn)
    return pd.DataFrame({'threshold':[thresh],
                         'tpr':[tpr], 
                         'fpr':[fpr]})
```
